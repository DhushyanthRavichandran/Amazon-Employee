```python
import numpy as np
import pandas as pd
!pip install catboost
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
test=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/amazon-employee-access-challenge/test.csv")
train=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/amazon-employee-access-challenge/train.csv")

x=train.drop(columns=['ACTION'],axis=1).values
y=train['ACTION']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33)
cat_features = [0, 1, 4, 5]
cb=CatBoostClassifier(learning_rate= 0.05)
cab=cb.fit(x_train,y_train,cat_features=cat_features)
pred=cb.predict(x_test)
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,pred)*100
acc
```

```python
prediction = cab.predict(x[0].reshape(1, -1))
prediction
```

```python
import os
import pickle

# Create the directory if it does not exist
os.makedirs('catboostModel', exist_ok=True)

# Open the file for writing
with open('catboostModel/catboostModel.pkl', 'wb') as f:
    pickle.dump(cab, f)

# Load the model from the pickle file
with open('catboostModel/catboostModel.pkl', 'rb') as f:
    pickleModel = pickle.load(f)

# Make a prediction using the loaded model
prediction = pickleModel.predict(x[0].reshape(1, -1))
prediction
```
